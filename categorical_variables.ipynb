## 绘制分类变量

```python
import pandas as pd
data = pd.read_csv("event_label_time.csv")

event = data.event.unique()
dict_event = {}
dict_event = {}.fromkeys([ev for ev in event])

event_num = data['event'].value_counts()

for key in dict_event.keys():
    dict_event[key] = event_num[key]

dict_event
```

```python
import matplotlib.pyplot as plt

names = list(dict_event.keys())
values = list(dict_event.values())

fig, axs = plt.subplots(1, 3, figsize=(15, 4), sharey=True)

axs[0].bar(names, values)
axs[1].scatter(names, values)
axs[2].plot(names, values)

axs[0].set_xticklabels(labels=names, rotation=90)
axs[1].set_xticklabels(labels=names, rotation=90)
axs[2].set_xticklabels(labels=names, rotation=90)
fig.suptitle('Categorical Plotting')
```
